# EDA

```python
train = pd.read_excel("data/raw/train(1).xlsx")
test  = pd.read_excel("data/raw/test2.xlsx")
```

```python
plt.figure(figsize=(8,4))
sns.histplot(train["price"], bins=50)
plt.title("Distribution of House Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

This is right skewed as very few properties are of premium prices this might lead to high variance so it is better to log transform it for better multi regression method application

Log tranformation

```python
train["log_price"] = np.log1p(train["price"])
plt.figure(figsize=(8,4))
sns.histplot(train["log_price"], bins=50)
plt.title("Log-Transformed Price Distribution")
plt.xlabel("log(price)")
plt.ylabel("Count")
plt.show()
```

Now the data is normally distributed.

# Univariate Analysis

```python
plt.figure(figsize=(6,4))
sns.countplot(x="bedrooms", data=train)
plt.title("Distribution of Bedrooms")
plt.show()
```

Most houses have 3-4 bedrooms.

Number of bathrooms also contain fractional values like 1.75,3.5 indicating for common bathrooms, luxurious bathrooms and attached bathrooms.

```python
plt.figure(figsize=(6,4))
sns.histplot(train["bathrooms"], bins=30)
plt.title("Distribution of Bathrooms")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(train["sqft_living"], bins=50)
plt.title("Distribution of Living Area (sqft)")
plt.show()
```

```python
num_features = [
    "sqft_living", "bathrooms", "bedrooms",
    "grade", "density_proxy", "sqft_living15"
]

plt.figure(figsize=(14,10))
for i, col in enumerate(num_features, 1):
    plt.subplot(3,2,i)
    sns.histplot(train[col], bins=40, kde=True)
    plt.title(f"Distribution of {col}")

plt.tight_layout()
plt.show()
```

```python
cat_features = ["waterfront", "view", "condition"]

plt.figure(figsize=(12,4))
for i, col in enumerate(cat_features, 1):
    plt.subplot(1,3,i)
    train[col].value_counts().plot(kind="bar")
    plt.title(f"{col} Frequency")

plt.tight_layout()
plt.show()
```

# Multivariate and Feature Analysis

```python
train["density_proxy"] = train["sqft_living15"] / train["sqft_lot15"]
test["density_proxy"] = test["sqft_living15"] / test["sqft_lot15"]
```

```python
sns.scatterplot(x="density_proxy", y="price", data=train, alpha=0.3)
plt.title("Neighborhood Density Proxy vs Price")
plt.show()
```

```python
corr_features = [
    "log_price", "bedrooms", "bathrooms",
    "sqft_living", "lat", "long","grade","waterfront","density_proxy","floors"
]

plt.figure(figsize=(8,6))
sns.heatmap(train[corr_features].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation with Log-Price (Key Features)")
plt.show()
```

```python
train_eda = train.copy()
train_eda.to_csv("data/processed/train_eda.csv", index=False)
test.to_csv("data/processed/test_eda.csv", index=False)
```

# Geospatial Analysis

```python
plt.figure(figsize=(6,6))
plt.scatter(train["long"], train["lat"], alpha=0.2)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Geographic Distribution of Properties")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(train["long"], train["lat"], c=train["price"], cmap="viridis", alpha=0.3)
plt.colorbar(label="Price")
plt.title("Spatial Distribution of Prices")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hexbin(
    train["long"],
    train["lat"],
    C=train["price"],
    gridsize=40,
    cmap="inferno",
    reduce_C_function=np.mean
)
plt.colorbar(label="Mean Price")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Geospatial Heatmap of Mean Property Price")
plt.show()
```

```python
train["price_bucket"] = pd.qcut(train["price"], q=3, labels=["Low", "Mid", "High"])

plt.figure(figsize=(8,6))
sns.scatterplot(
    x="long",
    y="lat",
    hue="price_bucket",
    data=train,
    alpha=0.5
)
plt.title("Spatial Distribution by Price Tier")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x="long",
    y="lat",
    hue="waterfront",
    data=train,
    alpha=0.6
)
plt.title("Spatial Distribution of Waterfront Properties")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(
    train["long"],
    train["lat"],
    c=train["sqft_living15"],
    cmap="viridis",
    alpha=0.4
)
plt.colorbar(label="Neighborhood Living Area (sqft_living15)")
plt.title("Neighborhood Scale Effect on Geography")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
